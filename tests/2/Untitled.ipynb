```python
import numpy as np
from sklearn.model_selection import LeaveOneOut
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
x = np.array([0, 4, 4.5, 3, 7, 8]).reshape(-1,1)
y = np.array([0, 0, 0,   1, 1, 1]).reshape(-1,1)
```

```python
clf = KNeighborsClassifier(n_neighbors=1)
loo = LeaveOneOut()
loo.get_n_splits(x)
```

```python
scores = []
for train_index, test_index in loo.split(x):
    train_x = x[train_index]
    train_y = y[train_index]
    print("[TRAIN]", train_x.ravel(), train_y.ravel())
    
    clf.fit(X=train_x, y=train_y.ravel())
    
    test_x = x[test_index]
    test_y = y[test_index]
    print("[TEST]", test_x, test_y)
    
    score = accuracy_score(y_true=test_y, y_pred=clf.predict(test_x))
    print("[SCORE]", score)
    scores.append(score)
    
np.mean(scores)
```

```python
from sklearn.neighbors import NearestCentroid

clf = NearestCentroid()
scores = []
for train_index, test_index in loo.split(x):
    train_x = x[train_index]
    train_y = y[train_index]
    print("[TRAIN]", train_x.ravel(), train_y.ravel())
    
    clf.fit(X=train_x, y=train_y.ravel())
    
    test_x = x[test_index]
    test_y = y[test_index]
    print("[TEST]", test_x, test_y)
    
    score = accuracy_score(y_true=test_y, y_pred=clf.predict(test_x))
    print("[SCORE]", score)
    scores.append(score)
    
np.mean(scores)
```

```python
X = list(np.array([[1,1],[3,-1],[1,-1]]))
w = np.array([0,0])

change = True
while(change):
    change = False
    for x in X:
        print("w=", w, "x=", x)
        prod = np.dot(x, w)
        if  prod <= 0:
            w += x
            change = True
        if prod > 0:
            w -= x
            change = True
print("w=", w)
```
