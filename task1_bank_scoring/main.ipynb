```python
import numpy as np
import pandas as pd
import sklearn
from tqdm import tqdm
import os
```

```python
train_path = "./train.csv"
test_path = "./test.csv"

train_data = pd.read_csv(train_path)
test_data = pd.read_csv(test_path)

train_data.head(10)
```

```python
train_data.fillna(0.0, inplace=True)
test_data.fillna(0.0, inplace=True)

train_x = train_data.values[:,1:]
train_y = train_data.values[:,0]

test_x = test_data.values
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import ShuffleSplit, cross_val_score
from sklearn.metrics import roc_auc_score

clf = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=1)
# clf.fit(train_x, train_y)
cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=20)
cross_val_score(clf, train_x, train_y, cv=cv, scoring="roc_auc")
```

```python
pred = clf.predict(train_x)
print(roc_auc_score(pred, train_y))

pd.DataFrame({'id': np.arange(test_x.shape[0]), 
              'a':clf.predict_proba(test_x)[:,1]}).to_csv('./solution.csv', index=False)
```
